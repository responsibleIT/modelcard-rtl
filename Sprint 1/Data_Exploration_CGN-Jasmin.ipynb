```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
```

```python
samenstelling_cgn = {
    "dialoog": {
        "conversaties": 3000000,
        "interviews": 460000,
        "telefoon-dialoog": 3000000,
        "zakelijke \n onderhandelingen": 175000,
        "interviews/ \n discussies": 750000,
        "lessen": 350000
    },
    "monoloog": {
        "beschrijvingen": 40000,
        "spontaan \n commentaar": 250000,
        "reportages":250000,
        "nieuws \n bulletins": 250000,
        "voorbereid \n commentaar": 200000,
        "lezingen": 275000,
        "voorgelezen \n tekst": 625000
    }
}

fig, axs = plt.subplots(2, 1)
fig.set_figwidth(10)
fig.set_figheight(10)
fig.suptitle("CGN word distribution")
fig.supylabel("WordCount")

axs[0].set_title("Dialoog")
axs[0].bar(range(len(samenstelling_cgn["dialoog"])), samenstelling_cgn["dialoog"].values())
axs[0].set_xticks(range(len(samenstelling_cgn["dialoog"])), samenstelling_cgn["dialoog"].keys())

axs[1].set_title("Monoloog")
axs[1].bar(range(len(samenstelling_cgn["monoloog"])), samenstelling_cgn["monoloog"].values())
axs[1].set_xticks(range(len(samenstelling_cgn["monoloog"])), samenstelling_cgn["monoloog"].keys())

fig.show()
```
